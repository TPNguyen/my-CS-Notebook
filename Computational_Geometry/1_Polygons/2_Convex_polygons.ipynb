# Convex polygons
## Definition
+ A polygon P is convex if
    + Given 2 points a,b inside P $(a,b \in P)$
    + P contains the entire segment **ab**
- Properties
    + interior angles of convex polygon always $ < \pi$

#### Non-convex polygon and reflex vertices
- Reflex vertex
    + The interior angle of that vetex $>\pi$
- Convex vertex
    + The interior angle of that vetex $<\pi$
- Degenerate vertex
    + Interior angle = Exterior angle $=\pi$
    + Degenerate vertices should be checked and eliminate

- How to solve Non-convex polygon problems  
    => [Convex partition] Divide into convex polygons and solve

## Algorithm: Check if a point is inclusion in a convex polygon
- Test inclusion of a point **q** in a convex polygon **P**
    + Time complexity
        + Preprocessing: O(n)
        + Query: O(log(n))
    + Space complexity: O(n)

### 1. Pick a point z strictly interior to P
- Often we choose point **z** =
    + center of massess from 3 non-collinear vetices of **P**: **u**, **v**, **w**
    + $z = \left( \frac{u_x+v_x+w_x}{3}, \frac{u_y+v_y+w_y}{3} \right)$

<img src="./assets/22.jpg" alt="drawing" width="150"/>

### 2. Construct rays from z
- for each vectex **v** of **P**
    + Generate a ray through **v** with origin at **z**

<img src="./assets/23.jpg" alt="drawing" width="150"/>

- The polygon **P** is partitioned into **n** wedges
    + 1 wedge has
        + 1 triangle area lying inside **P**
        + 1 edge of **P**
        + unbounded area lying outside **P**

<img src="./assets/24.jpg" alt="drawing" width="150"/>

### 3. Determine if point q inclusion in P
#### 3.1 Determine which wedge **q** belongs to in `O(log(n))`
+ Define the wedges = The angle between
    + The ray define the wedge and The positive horizontal direction ray
    + <img src="./assets/25.jpg" alt="drawing" width="150"/>
    + store and sort the angles into array **A**
        + We should store and maintain $\tan(angle)$ instead of the angle itself
+ Define angle $\alpha$ = angle between
    + the ray passing through **z** and **q** and The positive horizontal direction ray
    + <img src="./assets/26.jpg" alt="drawing" width="150"/>
    + Compute $\alpha$
        + $\tan(\alpha) = \frac{q_y - z_y}{q_x - z_x}$
        + We should store and maintain $\tan(\alpha)$ instead of $\alpha$
+ Use Binary search to determine which wedge **q** belongs to
    + Search for $\alpha$ in sorted array **A**

#### 3.2 Determine if **q** in/outside the wedge in `O(1)`
+ Define edge **e = (a,b)** of **P**
    + edge of the wedge contains **q**
+ <img src="./assets/27.jpg" alt="drawing" width="150"/>
+ **q** is inside **P** if and only if
    + **a**, **b**, **q** make a left turn
