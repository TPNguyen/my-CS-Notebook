## Binary Search Tree

```C++
class BinarySearchTree {
private:
    struct Node {
        int val;
        Node *left, *right;
        Node(int x) : val(x), left(NULL), right(NULL) {}
    };

    // pointer
    Node *root_;
public:
    BinarySearchTree() : root_(NULL) {}
    void insert(int val) {
        // Create new node
        Node *newNode = new Node(val);

        // Case empty tree
        if(root_ == NULL) {
            root_ = newNode;
            return;
        }

        // Find the correct position
        Node *cur = root_;
        while(cur != NULL) {
            // Insert left
            if(val <=  cur->val && cur->left == NULL) {
                cur->left = newNode;
                return;
            }

            // Insert right
            if(val > cur->val && cur->right == NULL) {
                cur->right = newNode;
                return;
            }

            // Traverse
            cur = (val <= cur->val) ? cur->left : cur->right;
        }
    }
private:
    void preOrder(Node *cur) {
        // Terminate condition
        if(cur == NULL) return;

        // preOrder Traverse
        cout << cur->val << ' ';
        preOrder(cur->left);
        preOrder(cur->right);
    }
public:
    void preOrderTraverse() {
        preOrder(root_);
    }
    void levelOrderTraverse() {
        // Check empty Tree
        if(root_ == NULL) return;

        // Init queue
        queue<Node*> Q;
        Q.push(root_);

        Node *cur = NULL;
        while(!Q.empty()) {
            // Pop queue
            cur = Q.front();
            Q.pop();

            // Process current Node
            cout << cur->val << ' ';

            // Enqueue next left, right
            if(cur->left != NULL) Q.push(cur->left);
            if(cur->right != NULL) Q.push(cur->right);
        }
    }
};
```
