## Binary knapsack
- N items:
    + weight[1:N]
    + value[1:N]

- Find the maximum value of combined items that not exceeded W


```C++
int W, N;
int wei[303];
int val[303];
```


## Solution

- `dfs` + Optimize the tree

<img src="./img/10.jpg" alt="drawing" width="800"/>



#### BottomUp - Dynamic Programming
```C++
// state: val = (i, w)
int dp[303][10003];
int solveDP() {
    // Init
    ms(dp, -1);
    dp[0][0] = 0;

    // DP
    FOR(i,0,N) FOR(w,0,W+1) {
        int cur = dp[i][w];
        if(cur == -1)
            continue;

        // Choose i+1
        if(w+wei[i+1] <= W)
            dp[i+1][w+wei[i+1]] = max(dp[i+1][w+wei[i+1]], cur + val[i+1]);

        // Not choose i+1
        dp[i+1][w] = max(dp[i+1][w], cur);
    }

    // Get results
    int res = -1;
    FOR(w,0,W+1) 
        res = max(res, dp[N][w]);
    return res;
}
```

#### BottomUp - Backtracking with DP Optimization


```C++
// state: val = (i, w)
int dp[303][10003];
int res;
void get(int i, int w) {
    int cur = dp[i][w];
    
    // Get results
    if(i == N) {
        res = max(res, cur);
        return;
    }

    // Choose i+1
    if(w+wei[i+1] <= W &&
            dp[i+1][w+wei[i+1]] < cur + val[i+1]) {
        dp[i+1][w+wei[i+1]] = cur + val[i+1];
        get(i+1, w+wei[i+1]);
    }

    // Not choose i+1
    if(i+1 <= N &&
            dp[i+1][w] < cur) {
        dp[i+1][w] = cur;
        get(i+1, w);
    }
}

int solveDP() {
    // Init
    ms(dp,-1);
    res = -1;

    // dfs - DP
    dp[0][0] = 0;
    get(0,0);
    return res;
}
```

#### Topdown
```C++
// state: val = (i, w)
int dp[303][10003];
int get(int i, int w) {
    if(i<=0 || w<=0) return 0;
    int &res = dp[i][w];
    if(res != -1) return res;

    // Choose i
    res = max(res, get(i-1, w-wei[i]) + val[i]);

    // Not choose i
    res = max(res, get(i-1, w));

    return res;
}
int solveDP() {
    ms(dp, -1);
    return get(N,W);
}
```
