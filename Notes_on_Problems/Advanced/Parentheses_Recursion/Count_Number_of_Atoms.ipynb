## [Number of Atoms](https://leetcode.com/problems/number-of-atoms/)

- Given a chemical formula (given as a string), return the count of each atom.
- An atomic element always starts with an uppercase character, then zero or more lowercase letters, representing the name.

#### Example
```
Input: 
formula = "H2O"
Output: "H2O"
Explanation: 
The count of elements are {'H': 2, 'O': 1}.


Input: 
formula = "Mg(OH)2"
Output: "H2MgO2"
Explanation: 
The count of elements are {'H': 2, 'Mg': 1, 'O': 2}.


Input: 
formula = "K4(ON(SO3)2)2"
Output: "K4N2O14S4"
Explanation: 
The count of elements are {'K': 4, 'N': 2, 'O': 14, 'S': 4}.
```

#### Solution

```C++
class Solution {
public:
    string s;
    int N;

    bool isCap(char c) {
        return 'A' <= c && c <= 'Z' ? true : false; 
    }
    bool isNor(char c) {
        return 'a' <= c && c <= 'z' ? true : false; 
    }
    bool isNum(char c) {
        return '0' <= c && c <= '9' ? true : false; 
    }

    map<string, int> parse(int &i) {
        map<string, int> res;

        while(i < N) {
            if(s[i] == '(') {
                ++i;

                // parse inside ()
                map<string, int> paren = parse(i);

                // merge hashmap
                for(auto &it: paren) res[it.first] += it.second;
            }

            else if(s[i] == ')') {
                ++i;

                // Get the multiplier
                int multi = 0;
                while(isNum(s[i])) {
                    multi = multi*10 + (s[i]-'0');
                    ++i;
                }

                // Update multiplier
                for(auto &it: res) res[it.first] *= multi;
                return res;
            }

            else if(isCap(s[i])) {
                // Get the elem
                string elem = "";
                elem.push_back(s[i]);
                ++i;

                while(isNor(s[i])) {
                    elem.push_back(s[i]);
                    ++i;
                }

                // Get the count
                int cnt = 0;
                while(isNum(s[i])) {
                    cnt = cnt*10 + (s[i]-'0');
                    ++i;
                }

                // Update hashmap
                cnt == 0 ? res[elem] += 1 : res[elem] += cnt; 
            }
        }

        return res;
    }
    string countOfAtoms(const string &formula) {
        s = formula;
        N = s.size();

        string ret = "";

        int idx = 0;
        map<string, int> ans = parse(idx);
        for(auto &it: ans) {
            ret += it.first;
            ret += it.second==1 ? "" : to_string(it.second);
        }
        return ret;
    }
};
```
