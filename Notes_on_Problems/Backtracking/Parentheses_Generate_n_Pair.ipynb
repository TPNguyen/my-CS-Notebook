## [Generate Parentheses](https://leetcode.com/problems/generate-parentheses/)

- Given n pairs of parentheses, write a function to generate all combinations of well-formed parentheses.
- Example:

```
n = 1
    ()

n = 2
    (())
    ()()

n = 3
    ((()))
    (()())
    (())()
    ()(())
    ()()()
```

#### Solution

```C++
class Solution {
private:
    int N;

    vector<string> ans;

    string s;
    void dfs(int open, int close) {
        if(s.size() == N*2) {
            ans.push_back(s);
            return;
        }

        // Case: generate (
        if(open < N) {
            s.push_back('(');
            dfs(open+1, close);
            s.pop_back();
        } 
        
        // Case: generate )
        if(close < open) {
            s.push_back(')');
            dfs(open, close+1);
            s.pop_back();
        }
    }
public:
    vector<string> generateParenthesis(int n) {
        N = n;
        dfs(0, 0);
        return ans;
    }
};
```
