
## [Median of 2 sorted arrays](https://leetcode.com/problems/median-of-two-sorted-arrays/)
- Example
```
    A = {-5, 3, 6, 12, 15}
    B = {-12, -10, -6, -3, 4, 10}
        => Median of Merge(A, B) = 3.0
```
```
    A = {2, 3, 5, 8}
    B = {10, 12, 14, 16, 18, 20}
        => Median of Merge(A, B) = (10 + 12) / 2 = 11.0
```

#### Solution  O(log (m+n))
- Partition 2 arrays into 2 halves
    + 1st half(Blue) = small numbers
    + 2nd half(Red) = large numbers

```C++
class Solution {
private:
    double findMedian(const vector<int> &A,const vector<int> &B) {
        int n = A.size();
        int m = B.size();

        // ---------------------- Partition ------------------ //
        int l = 0;
        int r = n;
        int i, j;
        while(l <= r) {
            i = l + (r-l) / 2;
            j = (n + m + 1) / 2 - i;

            // Case: --i, ++j -> Search new i in the 1st half of A
            if(i > 0 && j < m && A[i-1] > B[j])
                r = i-1;

            // Case: ++i, --j -> Search new i in the 2nd half of A
            else if(i < n && j > 0 && B[j-1] > A[i])
                l = i+1;

            // Found the split
            else 
                break;
        }

        // ---------------------- Calculate median ------------------ //
        // med1 = max 1st half
        // med2 = min 2nd half
        int med1, med2;

        if(i == 0)
            med1 = B[j-1];
        else if(j == 0)
            med1 = A[i-1];
        else
            med1 = max(A[i-1], B[j-1]);

        // ------ Case: Odd
        if((n+m)%2 != 0)
            return static_cast<double>(med1);

        // ------ Case: Even
        if(i == n)
            med2 = B[j];
        else if(j == m)
            med2 = A[i];
        else
            med2 = min(A[i], B[j]);

        return (static_cast<double>(med1) + static_cast<double>(med2)) / 2.0;
    }
public:
    double findMedianSortedArrays(const vector<int> &nums1,const vector<int> &nums2) {
        if(nums1.size() < nums2.size())
            return findMedian(nums1, nums2);
        return findMedian(nums2, nums1);
    }
};
```
