## [Parsing A Boolean Expression](https://leetcode.com/problems/parsing-a-boolean-expression/)
- Return the result of a boolean expression
- An expression can either be:
    - "t", evaluating to True;
    - "f", evaluating to False;
    - "!(expr)", evaluating to the logical NOT of the inner expression expr;
    - "&(expr1,expr2,...)", evaluating to the logical AND of 2 or more inner expressions expr1, expr2, ...;
    - "|(expr1,expr2,...)", evaluating to the logical OR of 2 or more inner expressions expr1, expr2, ...
- **Examples**

```
Input: expression = "!(f)"
Output: true

Input: expression = "|(f,t)"
Output: true

Input: expression = "&(t,f)"
Output: false

Input: expression = "|(&(t,f,t),!(t))"
Output: false
```

#### Solution

```C++
class Solution {
public:
    string S;
    int N;
    bool get(int &i, const char &func) {
        bool term=true;
        if(func == '|') term = false;
        else if(func == '&') term = true;

        while(i<N) {
            if(S[i]=='t') {
                ++i;

                if(func == '&') term &= true;
                else if(func == '|') term = true;
                else term = true;
            } else if(S[i]=='f') {
                ++i;

                if(func == '&') term = false;
                else if(func == '|') term |= false;
                else term = false;
            } else if(S[i]=='!') {
                i += 2;

                if(func == '&') term &= get(i, '!');
                else if(func == '|') term |= get(i, '!');
                else term = get(i, '!');
            } else if(S[i]=='&') {
                i += 2;

                if(func == '&') term &= get(i, '&');
                else if(func == '|') term |= get(i, '&');
                else term = get(i, '&');
            } else if(S[i]=='|') {
                i += 2;
                
                if(func == '&') term &= get(i, '|');
                else if(func == '|') term |= get(i, '|');
                else term = get(i, '|');
            } else if(S[i]==')') {
                ++i;

                if(func == '!') return !term;
                return term;
            } else ++i;
        }
        return term;
    }
    bool parseBoolExpr(const string &str) {
        this->S = str;
        this->N = S.size();

        int i = 0;
        return get(i, 'x');
    }
};
```
