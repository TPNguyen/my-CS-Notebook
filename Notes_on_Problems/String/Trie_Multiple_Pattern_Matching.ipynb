## [Index Pairs of a String](https://leetcode.com/problems/index-pairs-of-a-string/)

- Give a list of patterns and a text
- return all index pairs [i, j] so that the substring text[i]...text[j] is in the list of patterns.
- **Examples**

```
Input: 
    text = "thestoryofleetcodeandme", 
    patterns = ["story","fleet","leetcode"]
Output: [[3,7],[9,13],[10,17]]

Input: 
    text = "ababa", 
    patterns = ["aba","ab"]
Output: [[0,1],[0,2],[2,3],[2,4]]
Explanation: 
    Notice that matches can overlap, see "aba" is found in [0,2] and [2,4].
```

#### Bruteforce
```C++
class Solution {
public:
    vector<vector<int>> indexPairs(const string &text, const vector<string>& patterns) {
        vector<vector<int>> ans;

        // Exhaustive search
        for(auto &w: patterns) {
            int p = -1;
            do {
                p = text.find(w, p+1);
                if(p != -1) ans.push_back( {p, p+w.size()-1} );
            } while(p != -1);

        }
        sort(ans.begin(), ans.end());
        return ans;
    }
};
```

#### Trie

```C++
class Trie {
private:
    class Node {
    public:
        Node *children[26];
        bool isWord;
        int len;
    public:
        Node(int len) : len(len) {
            this->isWord = false;
            memset(this->children, NULL, sizeof(this->children));
        }
    };
    Node *root;
public:
    Trie() {
        root = new Node(0);
    }
    void insert(const string &P) {
        Node *cur = root;
        for(const char &c:P) {
            if(cur->children[c-'a'] == NULL) 
                cur->children[c-'a'] = new Node(cur->len + 1);
            cur = cur->children[c-'a'];
        }
        cur->isWord = true;
    }
    vector<vector<int>> search_index(const string &T) {
        vector<vector<int>> res;
        int idx;
        Node *cur;
        for(int i=0; i<T.size(); ++i) {
            cur = root;
            idx = i;
            while(true) {
                cur = cur->children[T[idx]-'a']; 
                if(cur == NULL) break;

                if(cur->isWord) res.push_back({idx - cur->len + 1, idx});

                ++idx;
                if(idx >= T.size()) break;
            }
        }
        return res;
    }
};
class Solution {
public:
    vector<vector<int>> indexPairs(string &text, vector<string>& patterns) {
        Trie *trie = new Trie;
        for(auto &s:patterns) trie->insert(s);
        return trie->search_index(text);
    }
};
```
