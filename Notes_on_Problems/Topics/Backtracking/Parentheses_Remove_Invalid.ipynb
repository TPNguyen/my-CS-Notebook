## [Remove Invalid Parentheses](https://leetcode.com/problems/remove-invalid-parentheses/)
- Remove the minimum number of invalid parentheses in order to make the input string valid. Return all possible results.
- Note: The input string may contain letters other than the parentheses ( and ).

- Example

```
Input: "()())()"
Output: ["()()()", "(())()"]

Input: "(a)())()"
Output: ["(a)()()", "(a())()"]

Input: ")("
Output: [""]
```

#### Solution

```C++
class Solution {
public:
    string s;
    int N;

    unordered_set<string> ans;

    string sol;
    void dfs(int x, int pair, int remove_left, int remove_right) {
        if(x == N) {
            if(pair==0 && remove_left==0 && remove_right==0) ans.insert(sol);
            return;
        }

        // Case '('
        if(s[x]=='('){
            // remove the left-half-parentheses
            if(remove_left > 0) dfs(x+1, pair, remove_left-1, remove_right);

            // keep the  left-half-parentheses
            sol.push_back('(');
            dfs(x+1, pair+1, remove_left, remove_right);
            sol.pop_back();
        }

        // Case ')'
        else if(s[x]==')'){
            // remove the right-half-parentheses
            if(remove_right > 0) dfs(x+1, pair, remove_left, remove_right-1);

            // keep the  right-half-parentheses
            if(pair > 0) {
                sol.push_back(')');
                dfs(x+1, pair-1, remove_left, remove_right);
                sol.pop_back();
            }
        }

        // Other character
        else{
            sol.push_back(s[x]);
            dfs(x+1, pair, remove_left, remove_right);
            sol.pop_back();
        }
    }

    vector<string> removeInvalidParentheses(const string &str) {
        s = str;
        N = s.size();

        int remove_left = 0;
        int remove_right = 0;
        for(const char &c:s) {
            if(c == '(') ++remove_left;
            else if(c == ')'){
                if(remove_left > 0) --remove_left;
                else ++remove_right;
            }
        }

        dfs(0, 0, remove_left, remove_right);
        return vector<string>(ans.begin(), ans.end());
    }
};
```
