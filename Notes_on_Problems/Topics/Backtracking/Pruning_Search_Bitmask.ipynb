## [Smallest Sufficient Team](https://leetcode.com/problems/smallest-sufficient-team/)
- Given:
    + `req_skills`: list of required skills
    + `people`: list of people
- Find
    - `smallest sufficient team`: 
        + a set of people such that for every required skill in req_skills 
        + smallest possible size, represented by the index of each person.

- Constraints
    + 1 <= req_skills.length <= 16
    + 1 <= people.length <= 60
- Examples

```
Input: 
    req_skills = ["java","nodejs","reactjs"],
    people = [["java"],["nodejs"],["nodejs","reactjs"]]
Output: [0,2]

Input: 
    req_skills = ["algorithms","math","java","reactjs","csharp","aws"],
    people = [["algorithms","math","java"],["algorithms","math","reactjs"],["java","csharp","aws"],["reactjs","csharp"],["csharp","math"],["aws","java"]]
Output: [1,2]


```



#### Solution - integer bitset - 824 ms	
```C++
class Solution {
public:
    vector<int> A;
    int N, K;
    vector<int> ans;

    vector<int> res;
    void dfs(int x, int cur) {
        if(cur == (1 << K) - 1) { // Relax      
            if(ans.size() > res.size()) ans = res;
            return;
        }
        if(res.size() >= ans.size()) return;  // Pruning
        if(x == N) return;

        /* ------- Pruning ------- */
        // if add A[x] not increase the skill pool
        // Not choose x
        if((cur | A[x]) == cur) {
            dfs(x+1, cur);
            return; 
        }

        /* ------- Consider A[x] ------- */
        // Choose x
        res.push_back(x);
        dfs(x+1, cur | A[x]);
        res.pop_back();

        // Not choose x
        dfs(x+1, cur);
    }
    vector<int> smallestSufficientTeam(
            vector<string>& req_skills, 
            vector<vector<string>>& people) {
        // string to int
        K = req_skills.size();
        unordered_map<string, int> string_to_int;
        for(int i=0; i<K; ++i) string_to_int[req_skills[i]] = i;

        // setup bitset
        N = people.size();
        for(auto &p:people) {
            int cur=0;
            for(string &skill:p) cur |= (1 << string_to_int[skill]);
            A.push_back(cur);
        }

        // dfs
        ans.assign(17, 0);
        dfs(0, 0);
        return ans;
    }
};
```

#### Solution - bitset - 1348 ms
```C++
class Solution {
public:
    vector<bitset<17>> A;
    int N, K;
    vector<int> ans;

    vector<int> res;
    void dfs(int x,const bitset<17> &cur) {
        if(cur.count() == K) { // Relax      
            if(ans.size() > res.size()) ans = res;
            return;
        }
        if(res.size() >= ans.size()) return;  // Pruning
        if(x == N) return;

        /* ------- Pruning ------- */
        // if add A[x] not increase the skill pool
        // Not choose x
        if((cur | A[x]) == cur) {
            dfs(x+1, cur);
            return; 
        }

        /* ------- Consider A[x] ------- */
        // Choose x
        res.push_back(x);
        dfs(x+1, cur | A[x]);
        res.pop_back();

        // Not choose x
        dfs(x+1, cur);
    }
    vector<int> smallestSufficientTeam(
            vector<string>& req_skills, 
            vector<vector<string>>& people) {
        // string to int
        K = req_skills.size();
        unordered_map<string, int> string_to_int;
        for(int i=0; i<K; ++i) string_to_int[req_skills[i]] = i;

        // setup bitset
        N = people.size();
        for(auto &p:people) {
            bitset<17> cur=0;
            for(string &skill:p) cur |= (1 << string_to_int[skill]);
            A.push_back(cur);
        }

        // dfs
        ans.assign(17, 0);
        bitset<17> cur=0;
        dfs(0, cur);
        return ans;
    }
};
```

#### Solution - Sorting - 484 ms

```C++
class Solution {
public:
    struct People {
        int idx;
        int skill_set;
        int num_skill;
    };
    vector<People> A;
    int N, K;
    vector<int> ans;

    vector<int> res;
    void dfs(int x, int cur) {
        if(cur == (1 << K) - 1) { // Relax      
            if(ans.size() > res.size()) ans = res;
            return;
        }
        if(res.size() >= ans.size()) return;  // Pruning
        if(x == N) return;

        for(int i=x; i<N; ++i) {
            /* ------- Pruning ------- */
            // if add A[x] not increase the skill pool
            // Not choose x
            if((cur | A[i].skill_set) == cur) {
                dfs(i+1, cur);
                return; 
            }

            /* ------- Consider A[x] ------- */
            res.push_back(A[i].idx);
            dfs(i+1, cur | A[i].skill_set);
            res.pop_back();
        }
    }
    vector<int> smallestSufficientTeam(
            vector<string>& req_skills, 
            vector<vector<string>>& people) {
        // string to int
        K = req_skills.size();
        unordered_map<string, int> string_to_int;
        for(int i=0; i<K; ++i) 
            string_to_int[req_skills[i]] = i;

        // setup bitset
        N = people.size();
        for(int i=0; i<N; ++i) {
            int skill_set=0;
            for(string &skill:people[i]) 
                skill_set |= (1 << string_to_int[skill]);
            A.push_back( {i, skill_set, people[i].size()} );
        }

        // dfs
        ans.assign(17, 0);
        sort(A.begin(), A.end(), 
            [](People &x, People &y) {return x.num_skill > y.num_skill;});
        dfs(0, 0);
        return ans;
    }
};
```
