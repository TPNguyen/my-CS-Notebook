## [Different Ways to Add Parentheses](https://leetcode.com/problems/different-ways-to-add-parentheses/)
- Given a string of numbers and operators, return all possible results from computing all the different possible ways to group numbers and operators. 
- The valid operators are +, - and *.
- **Examples**

```
Input: "2-1-1"
Output: [0, 2]
Explanation: 
    ((2-1)-1) = 0 
    (2-(1-1)) = 2


Input: "2*3-4*5"
Output: [-34, -14, -10, -10, 10]
Explanation: 
    (2*(3-(4*5))) = -34 
    ((2*3)-(4*5)) = -14 
    ((2*(3-4))*5) = -10 
    (2*((3-4)*5)) = -10 
    (((2*3)-4)*5) = 10
    
    
Input: "11+69*3"
Output: [218, 240]
Explanation: 
    11+(69*3) = 218
    (11+69)*3 = 240
```

#### Solution

```C++
class Solution {
public:
    vector<string> a;
        // num idx = even: 0 2 4 6 ...
        // op  idx = odd: 1 3 5 ... 

    vector<int> get(int i, int j) {
        if(i>j) return vector<int>();
        if(i==j && i%2==0) return vector<int>({stoi(a[i])});

        // k = operators in [i,j]. 
        //   Eg: [i,j] = (num1 + num2 - num3), k = {+, -}
        vector<int> left, right;
        vector<int> res;
        for(int k=i+1; k<j; k+=2) {
            left = get(i,k-1);
            right = get(k+1,j);

            for(auto &l:left) for(auto &r:right) {
                if(a[k] == "+") res.push_back(l+r);
                else if(a[k] == "-") res.push_back(l-r);
                else if(a[k] == "*") res.push_back(l*r);
            }
        }
        return res;
    }
    vector<int> diffWaysToCompute(const string &input) {
        // Preprocess a = nums and ops
            // num idx = even: 0 2 4 6 ...
            // op  idx = odd: 1 3 5 ... 
        // Example: 11+69*3 --> {11, +, 69, *, 3}
        string num, op;
        for(auto &c:input) {
            if(c=='+' || c=='-' || c=='*') {
                a.push_back(num);
                num = "";

                op = c;
                a.push_back(op);
            }
            else num.push_back(c);
        }
        a.push_back(num);

        return get(0,a.size()-1);
    }
};
```
