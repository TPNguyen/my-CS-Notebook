## [Coin sums](https://projecteuler.net/problem=31)
- In England the currency is made up of pound, £, and pence, p, and there are eight coins in general circulation:
    + 1p, 2p, 5p, 10p, 20p, 50p, £1 (100p) and £2 (200p).

- How many different ways can £2 be made using any number of coins?
    + An example of configuration: 1×£1 + 1×50p + 2×20p + 1×5p + 1×2p + 3×1p


#### Backtrack: 2813044 loops

```C++
int cnt;
vector<int> coins = {1,2,5,10,20,50,100,200};
int NUM_COINS = coins.size();
int AMOUNT = 200;

void dfs(int x, int idx) {
    if(x > AMOUNT) return;
    if(x == AMOUNT) {
        ++cnt;
        return;
    }

    // Add idx to avoid same configurations
    for(int i=idx; i<NUM_COINS; ++i) dfs(x+coins[i], i);
}

void solve() {
    cnt = 0;
    sort(coins.begin(), coins.end());

    dfs(0, 0);
    cout<<cnt;
}
```

#### DP: 291 loops

```C++
vector<int> coins = {1,2,5,10,20,50,100,200};

int dp[203][11];
int get(int x, int idx) {
    if(x == 0 && idx < 0) return 1;
    if(x < 0 || idx < 0) return 0;
    if(dp[x][idx] > 0) return dp[x][idx];

    // idx = index of the current considering coin
    int amount = x;
    while(amount >= 0) {
        dp[x][idx] += get(amount, idx-1);
        amount -= coins[idx];
    }

    return dp[x][idx];
}

void solve() {
    memset(dp, 0, sizeof(dp));
    cout<<get(200, 7);
}
```

##### Mixed DP

```C++
vector<int> coins = {1,2,5,10,20,50,100,200};
int NUM_COINS = coins.size();
int AMOUNT = 200;

int dp[203];
void solve() {
    memset(dp, 0, sizeof(dp));
    
    dp[0] = 1;
    for(int idx=0; idx<NUM_COINS; ++idx) {
        for(int x=coins[idx]; x<=AMOUNT; ++x) {
            dp[x] += dp[x-coins[idx]];
        }
    }
    cout<<dp[AMOUNT];
}
```
